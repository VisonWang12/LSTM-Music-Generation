```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
# load audio files with librosa
scale, sr = librosa.load("LSTM_music.wav")
```

```python
# Extract mel spectrogram
mel_spectrogram = librosa.feature.melspectrogram(scale, sr=sr, n_fft=2048, hop_length=512, n_mels=50)
mel_spectrogram.shape
```

```python
# Visualize log-mel spectrogram
log_mel_spectrogram = librosa.power_to_db(mel_spectrogram)
plt.figure(figsize=(18, 5))
librosa.display.specshow(log_mel_spectrogram, 
                         x_axis="time",
                         y_axis="mel", 
                         sr=sr)
plt.colorbar(format="%+2.f")
plt.title('Log-Mel Spectrogram', fontsize=10)
plt.show()
```

```python
# Visualize zero crossing rate

zcrs = librosa.feature.zero_crossing_rate(scale + 0.0001)
plt.figure(figsize=(18, 5))
plt.plot(zcrs[0])
plt.title('Zero Crossing Rate')
```

```python
chroma = librosa.feature.chroma_cqt(y=scale, sr=sr)
plt.figure(figsize=(18, 5))
img = librosa.display.specshow(chroma, y_axis='chroma', x_axis='time')
plt.colorbar(img)
plt.title('Chromagram')
```
